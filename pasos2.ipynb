# 🧠 Proyecto: Detección de Tumores con Red Neuronal

Este proyecto tiene como objetivo desarrollar un modelo de red neuronal capaz de predecir si un tumor es benigno o maligno a partir de un conjunto de datos médicos.

Se utilizará Python 3.11.5 y TensorFlow/Keras siguiendo buenas prácticas de Machine Learning.

# 🎯 Objetivos del Proyecto

- Construir una red neuronal funcional usando TensorFlow/Keras
- Predecir correctamente tumores benignos y malignos
- Evaluar el modelo con métricas adecuadas (accuracy, precision, recall, F1-score)
- Minimizar errores en la detección de tumores malignos (alta prioridad)
- Aplicar buenas prácticas de Machine Learning
- Generar un modelo confiable y bien interpretado

# 🚨 Errores que se deben evitar

❌ No usar TensorFlow (usar sklearn en su lugar)
❌ Mezclar modelos (Keras + MLPClassifier)
❌ Usar solo accuracy como métrica
❌ No analizar los datos antes de entrenar
❌ No normalizar los datos
❌ Ignorar el overfitting
❌ No usar validación
❌ No interpretar resultados
❌ Obtener accuracy menor a 0.5 (modelo inútil)
❌ No revisar si TensorFlow está correctamente instalado

⚠️ IMPORTANTE:
Un modelo con bajo recall en cáncer es peligroso porque no detecta tumores malignos.
